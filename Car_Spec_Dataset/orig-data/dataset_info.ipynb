```python
string_columns = df.select_dtypes(include=['object'])

# Function to find length of longest string in a column, skipping non-string values
def longest_string_length(column):
    string_lengths = column.apply(lambda x: len(x) if isinstance(x, str) else 0)
    return max(string_lengths)

# Find length of longest string value for each string column
longest_lengths = string_columns.apply(longest_string_length)
print(longest_lengths)
```
