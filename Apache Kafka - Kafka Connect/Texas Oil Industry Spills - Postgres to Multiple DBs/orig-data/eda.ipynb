```python
import pandas as pd
```

```python
def longest_value_length(df):
    # Convert all values in the DataFrame to strings
    df_str = df.astype(str)
    
    # Apply a function to each column to find the maximum length of the string values
    max_lengths = df_str.applymap(len).max()
    
    return max_lengths
```

### `Central` Dataset is up First!

```python
df = pd.read_csv('central_cleaned.csv')

df
```

```python
print(longest_value_length(df))
```

### `District` Dataset is up Next!

```python
df = pd.read_csv('district_cleaned.csv')

df
```

```python
pd.set_option('display.max_columns', None)

print(longest_value_length(df).to_string())
```
