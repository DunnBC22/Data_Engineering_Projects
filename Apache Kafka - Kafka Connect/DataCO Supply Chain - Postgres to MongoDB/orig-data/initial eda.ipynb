```python
import pandas as pd

def max_value_length(df: pd.DataFrame) -> pd.Series:
    """
    Retrieves the length of the maximum value in each column of the DataFrame.
    
    Parameters:
    df (pd.DataFrame): The input DataFrame.
    
    Returns:
    pd.Series: A Series containing the length of the maximum value in each column.
    """
    # Convert all values in the DataFrame to strings
    df = df.astype(str)
    
    # Define a function to get the length of the maximum string value in a column
    def max_length(col):
        if col.empty:
            return 0
        max_value = col.max()
        return len(max_value)
    
    # Apply the function to each column
    lengths = df.apply(max_length)
    return lengths
```
